##### 2.- Crea dos archivos. Uno almacena los accession que empiezen por 'a' y el otro el resto

```python
accs = ['ab56', 'bh84', 'hv76', 'ay93', 'ap97', 'bd72']
```

```python
with open("accs_a.txt", "w") as file_a, open("accs_otros.txt", "w") as file_otros:
    for acc in accs:
        if acc.startswith('a'):
            file_a.write(acc + '\n')
        else:
            file_otros.write(acc + '\n')
```

```python
print("Archivos creados bien:")
print(" - accs_a.txt contiene los que empiezan con 'a'")
print(" - accs_otros.txt contiene el resto")
```

##### 4.- Escribe una función que cuente el porcentaje de A de una secuencia

```python
def porcentaje_A(secuencia):
    secuencia = secuencia.upper()
    
    num_A = secuencia.count('A')
    
    porcentaje = (num_A / len(secuencia)) * 100
    
    return porcentaje

seq = "ATGCAAACTGA"
print(f"Porcentaje de A: {porcentaje_A(seq):.2f}%")
```

##### ¿Puedes hacer que devuelva dos decimales? ¿Y que lidie con mayúsculas y minúsculas?

```python
def porcentaje_A(secuencia):
    if len(secuencia) == 0:
        return 0.0
    
    secuencia = secuencia.upper()
    
    num_A = secuencia.count('A')
    
    porcentaje = (num_A / len(secuencia)) * 100
    
    return round(porcentaje, 2)

seq = "AtGcAAactGA"
print(f"Porcentaje de A: {porcentaje_A(seq)}%")
```

##### ¿Y que devuelva el número de decimales que yo quiero?

```python
def porcentaje_A(secuencia, decimales=2):
    if len(secuencia) == 0:
        return 0.0

    secuencia = secuencia.upper()

    num_A = secuencia.count('A')

    porcentaje = (num_A / len(secuencia)) * 100

    return round(porcentaje, decimales)

seq = "AtGcAAactGA"
print("2 decimales:", porcentaje_A(seq, 2))
print("4 decimales:", porcentaje_A(seq, 4))
```

##### ¿Sabes poner un valor por defecto en un argumento?

```python
def porcentaje_A(secuencia, decimales=2):
    if len(secuencia) == 0:
        print("La secuencia está vacía.")
        return

    secuencia = secuencia.upper()

    num_A = secuencia.count('A')

    porcentaje = (num_A / len(secuencia)) * 100

    print(f"Porcentaje de A: {porcentaje:.{decimales}f}%")
```

```python
porcentaje_A("AtGcAAactGA")
porcentaje_A("AtGcAAactGA", 4)
```

##### ¿Qué pasa si no ponemos un argumento? ¿Sabrías hacer una función encapsulada?

```python
def porcentaje_A_obligatorio(secuencia, decimales=2):
    secuencia = secuencia.upper()
    if len(secuencia) == 0:
        print("La secuencia está vacía.")
        return
    num_A = secuencia.count('A')
    porcentaje = (num_A / len(secuencia)) * 100
    print(f"Porcentaje de A: {porcentaje:.{decimales}f}%")
```

```python
porcentaje_A_obligatorio("AtGcAAactGA", 2)
```

#### 5.- De la lista de accessions que se proporciona obtén lo siguiente:
1. Los que tengan el número 5
2. Los que tengan la letra 'd' o la 'e'
3. Los que tengan las letras 'd' y 'e' en ese orden
4. Los que tengan las letras 'd' y 'e' en ese orden con una letra cualquiera en medio
5. Los que tengan ambas letras 'd' y 'e' en cualquier orden
6. Los que empiezan por 'x' o por 'y'
7. Los que empiezan por 'x' o 'y' y terminan por 'e'
8. Los que tienen tres o más dígitos seguidos
9. Los que terminan con 'd' seguidos de 'a' o 'r' o 'p'

```python
import re
```

```python
accs = ['xkn59438', 'yhdck2', 'eihd39d9', 'chdsye847', 'hedle3455', 'xjhd53e', '45da', 'de37dp']
```

```python
tengan_5 = [a for a in accs if '5' in a]
print("1. Tienen el número 5:", tengan_5)
```

```python
tengan_d_o_e = [a for a in accs if re.search(r'[de]', a)]
print("2. Tienen 'd' o 'e':", tengan_d_o_e)
```

```python
d_y_e = [a for a in accs if re.search(r'd.*e', a)]
print("3. Tienen 'd' y 'e' en ese orden:", d_y_e)
```

```python
d_cualquiera_e = [a for a in accs if re.search(r'd.e', a)]
print("4. 'd' y 'e' con una letra en medio:", d_cualquiera_e)
```

```python
d_e = [a for a in accs if re.search(r'd', a) and re.search(r'e', a)]
print("5. Tienen 'd' y 'e' en cualquier orden:", d_e)
```

```python
x_o_y = [a for a in accs if re.match(r'^[xy]', a)]
print("6. Empiezan por 'x' o 'y':", x_o_y)
```

```python
empiezan_x_o_y_terminan_e = [a for a in accs if re.match(r'^[xy].*e$', a)]
print("7. Empiezan por 'x' o 'y' y terminan por 'e':", empiezan_x_o_y_terminan_e)
```

```python
tres_o_mas = [a for a in accs if re.search(r'\d{3,}', a)]
print("8. Tienen tres o más dígitos seguidos:", tres_o_mas)
```

```python
termina_d_seguido_de_a_r_p = [a for a in accs if re.search(r'd[arp]$', a)]
print("9. Terminan con 'd' seguido de 'a', 'r' o 'p':", termina_d_seguido_de_a_r_p)
```
